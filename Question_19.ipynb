#### Q19: GUI - Password Strength Checker

```python
from breezypythongui import EasyFrame
class PasswordStrength(EasyFrame):
    def __init__(self):
        EasyFrame.__init__(self, title="Password Strength Checker")

        self.addLabel(text="Enter Password:", row=0, column=0)
        self.passwordField = self.addTextField(text="", row=0, column=1)
        self.passwordField["show"] = "*"

        self.addButton(text="Check Strength",row=1, column=0, columnspan=2,
                       command=self.checkStrength)

        self.addLabel(text="Strength:", row=2, column=0)
        self.resultField = self.addTextField(text="", row=2, column=1)
        self.resultField["state"] = "readonly"

    def checkStrength(self):
        password = self.passwordField.getText()

        hasDigit = any(char.isdigit() for char in password)
        specialChars = "!@#$%^&*(),.?\":{}|<>"
        hasSpecial = any(char in specialChars for char in password)

        if 1<=len(password) < 8:
            strength = "Weak"
        elif hasDigit and hasSpecial:
            strength = "Strong"
        elif password == "":
            strength="Empty password"
        else:
            strength = "Moderate"

        self.resultField["state"] = "normal"
        self.resultField.setText(strength)

if __name__ == "__main__":
    PasswordStrength().mainloop()
```
