#### Q17: GUI - Login system

```python
from breezypythongui import EasyFrame
class Login(EasyFrame):
    def __init__(self):
        EasyFrame.__init__(self,title="Login system")

        self.addLabel(text="Username",row=0,column=0)
        self.usernameField=self.addTextField(text="",row=0,column=1)

        self.addLabel(text="Password",row=1,column=0)
        self.passwordField=self.addTextField(text="",row=1,column=1)
        self.passwordField["show"]="*"

        self.loginButton=self.addButton(text="Login",row=2,column=0,columnspan=2,command=self.checkLogin)

        self.correctUsername="admin"
        self.correctPassword="1234"

    def checkLogin(self):
        username=self.usernameField.getText()
        password=self.passwordField.getText()

        if username=="" or password=="":
            self.messageBox(title="Input is empty",message="Username or password is empty")
        elif username==self.correctUsername and password==self.correctPassword:
             self.messageBox(title="Logged In",message="Login successful!")

        

if __name__=="__main__":
    Login().mainloop()
```
